```python

# Define your prediction modes
Pred_mode = ["vp", "linear0.25", "linear0.5", "linear1"]
t = th.linspace(0.000, 1-1e-4, 80)
T = [t, t, t, th.linspace(0.000, 80, 80)]
# Create a figure and subplots
plt.figure(figsize=(8, 1.7), dpi=300)

for i, pred_mode in enumerate(Pred_mode):
    plt.subplot(1, 4, i+1)  # Create subplots
    # t = T[i]
    alpha, _, beta, _, gamma, _ = get_route(pred_mode)  # Get the routes for each pred_mode
    
    # Plot alpha, beta, gamma
    plt.plot(t, alpha(t), label="$\\alpha_t$")
    plt.plot(t, beta(t), label="$\\beta_t$")
    plt.plot(t, gamma(t), label="$\\gamma_t$")
    
    # Set ticks to display only 0 and 1
    # plt.xticks([0, 1])
    # plt.yticks([0, 1])
    
    # Add the legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Adjust the layout to fit the plots and legends nicely
plt.tight_layout()

```
